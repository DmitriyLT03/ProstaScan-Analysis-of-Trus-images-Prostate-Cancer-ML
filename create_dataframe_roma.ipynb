```python
folders_path = os.listdir('../all_data/Новая партия/')
```

```python
from pathlib import Path

dst = Path("../all_data/new_img")

def convert_and_save(src, dst):
    print(os.path.exists(src), os.path.exists(dst))
    ds = dicom.dcmread(src)
    pixel_array_numpy = ds.pixel_array
    st = cv2.imwrite(filename=dst, img=pixel_array_numpy)
    print("Saved:", dst, st)
        
        
for root, dirs, files in os.walk(path):
    if not files:
        continue
    
    folder_name = root.split("/")[-1]
    save_path = dst.joinpath(folder_name)
    save_path.mkdir(parents=True, exist_ok=True)
    for file in files:
        _src = Path(root, file)
        _dst = dst.joinpath(folder_name, file + ".jpg")
        convert_and_save(_src.__str__(), _dst.__str__())
        # convert_and_save(root + "/" + file, dst.joinpath(file + ".jpg").as_posix())
```

```python
for num, folder in enumerate(folders_path):
    folder_dir_save = os.path.join(jpg_folder_path, folder)
    if not os.path.exists(folder_dir_save):
            os.mkdir(folder_dir_save)
    folder_dir_path = os.path.join(path, folder)
    folder_dir = os.listdir(os.path.join(path, folder))
    for image in folder_dir:
        ds = dicom.dcmread(os.path.join(folder_dir_path, image))
        pixel_array_numpy = ds.pixel_array
        image = image.replace('.dcm', '.jpg')
        print(os.path.join(folder_dir_save, image) + '.jpg')
        cv2.imwrite(os.path.join(folder_dir_save, image)+'.jpg', pixel_array_numpy)
        
```
